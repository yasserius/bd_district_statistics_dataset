```python
import pickle
import pandas as pd
import glob
import os
import string
from pprint import pprint
```

```python
pkl_files = glob.glob("./pickles3/*.pkl")

print(len(pkl_files))
```

```python
print(pkl_files[0])
```

```python
def fix(input):
    output = ''
    
    for x in input:
        if x in string.printable:
            output += x
        else:
            output += " "
    
    return output.replace("/", ", ").replace("\t", " ").replace("\n", " ")
```

```python
def clean_table(input):
    output = []
    for row in input:
        new_row = [cell.replace('\n', ' ') for cell in row]
        output.append(new_row)
    
    return output
```

```python
def count_tables(blocks):
    c = 0
    for b in blocks:
        if b[0] == 't': c += 1
            
    return c
```

```python
for file in pkl_files:
    counter = {
        "type 1": 0,
        "type 2": 0,
        "actual tables": 0
    }
    
    district = file.split('\\')[-1].split('.')[0]
    print(district)
    
    export_dir = './CSVs_2/{}'.format(district)
    try:
        os.mkdir(export_dir)
    except:
        pass
        
    with open(file, 'rb') as f:
        data = pickle.load(f)
    
    for k, v in data.items():
        chapter = k
        
        counter["actual tables"] += count_tables(v)
        
        i = 0
        while i < len(v):
            if v[i][0] == 't' and \
            v[i-1][1] == 'List Paragraph':
                header = fix(v[i-1][2])
                subheader = "" # intentional blank
                
                data_table = clean_table(v[i][1])
                table = [[header], [subheader]] + data_table
                
                df = pd.DataFrame(table)
                df.to_csv(export_dir + '/{}_{}.csv'.format(chapter, header))
                counter["type 1"] += 1
                
            elif v[i][0] == 't' and \
            v[i-1][1] == 'Normal'and \
            v[i-2][1] == 'List Paragraph':
                header = fix(v[i-2][2])
                subheader = fix(v[i-1][2])
                
                data_table = clean_table(v[i][1])
                table = [[header], [subheader]] + data_table
                
                df = pd.DataFrame(table)
                df.to_csv(export_dir + '/{}_{}.csv'.format(chapter, header))
                counter["type 2"] += 1
                
            i += 1
    
    counter['total generated'] = counter['type 1'] + counter['type 2']
    pprint(counter)
    
    with open(export_dir + "/counts.txt", 'w') as f:
        f.write("type 1: " + str(counter['type 1']) + "\n")
        f.write("type 2: " + str(counter['type 2']) + "\n")
        f.write("total generated: " + str(counter['total generated']) + "\n")
        f.write("actual tables: " + str(counter['actual tables']) + "\n")
    
```
