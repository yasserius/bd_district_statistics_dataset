```python
import glob
import pandas as pd

from pprint import pprint
```

```python
csv_dir = "./verified_CSVs/*"

zilas = glob.glob(csv_dir)

print(zilas[:10])
```

```python
for zila in zilas:
    zila_name = zila.split('\\')[-1]
    
    output_html = """
        <head>
          <style>
          table {
            font-family: Arial, Helvetica, sans-serif;
            border-collapse: collapse;
            width: 100%;
          }

          table td, table th {
            border: 1px solid #ddd;
            padding: 8px;
          }

          table tr:nth-child(even){background-color: #f2f2f2;}

          #customers tr:hover {background-color: #ddd;}

          table th {
            padding-top: 12px;
            padding-bottom: 12px;
            text-align: left;
            background-color: #4CAF50;
            color: white;
          }
          </style>
        </head>
        <body>"""
    
    output_html +=  "<h1>{}</h1><br>".format(zila_name)
    
#     print('\n\n\nZila:', zila.split('\\')[-1])
    csv_files = glob.glob(zila + '/*.csv')
    
    if csv_files:
        for f in csv_files:
            df = pd.read_csv(f, index_col=0, encoding='cp1252')

            folder_name = f.replace('CSVs_4', 'cleaned_CSVs')

            output_html += '\n<br>\n' + \
                            '<h2>{}</h2>\n'.format(folder_name) + \
                            df.to_html(na_rep='')
    else:
        print("No CSV files in", zila)
        continue
    
    output_html += "</body>"
    
    
    export_dir = zila + "/tables.html"
    with open(export_dir, "w", encoding='cp1252') as f:
        try:
            f.write(output_html)
            print("HTML file created:", export_dir)
        except:
            print('Failed', zila)
```
